```python
import pandas as pd

fileName = 'BrazilCOVIDData.xlsx'
xls = pd.ExcelFile(fileName)
BrazilCovid_df = pd.read_excel(xls, 'Brazil Covid-19 data')
TempByState_df = pd.read_excel(xls, 'Temperature by State')
#StateStats_df = pd.read_excel(xls, 'Brazil State Stats')
CityArea_df = pd.read_excel(xls, 'City area')
```

```python
#renaming to appropriate column names
TempByState_df_2 = TempByState_df.rename(columns= {'ANNUAL': 'Temp', 'STATE_ABBR': 'State', 'CITY': 'City'})
BrazilCovid_df_2 = BrazilCovid_df.rename(columns= {'Population as of 2019': 'Population', 'Municipality': 'City'})
CityArea_df_2 = CityArea_df.rename(columns= {'ST': 'State'})



```

```python
BrazilCovid_df
```

```python
TempByState_df.head()
```

```python
joined_df = BrazilCovid_df_2.merge(TempByState_df_2, on=['City', "State"])
joined_df
```

```python
joined_df_2 = joined_df.merge(CityArea_df_2, on=['City', "State"])
joined_df_2
```

```python
#filtering down data
joined_df_2 = joined_df_2[['State', 'City', 'Population', 'SQ_KM', 'Accumulated cases', 'Temp' ]]
joined_df_2
```

```python
#checking the data types
joined_df_2.info()
```

```python
#changing pop to an int
joined_df_2[['Population']] = joined_df_2[['Population']].apply(pd.to_numeric) 
```

```python
#creating population density column
joined_df_2['Pop_dense'] = joined_df_2['Population']/joined_df_2["SQ_KM"]
joined_df_2
```

```python
#making a list to use as the days column
my_list = list(range(0,151)) * 27
```

```python
# turning list into df to convert to days column
my_list = pd.DataFrame(my_list)
my_list
```

```python
#addiding the days column to the df using the list
new_df = joined_df_2.assign(Days = my_list)
```

```python
new_df
```

```python
new_df.shape
```

```python
new_df.drop_duplicates(subset=['STATE_ABBR', 'Capitol City'])
```

```python
#adding features
new_df['days_cube'] = new_df['Days']**3
new_df['days_sq'] = new_df['Days']**2
new_df['Pop_dens_sq'] = new_df['Pop_dense']**2
```

```python
#checking it out with all the features
new_df
```

```python
# Get Final Features DataFrame
features = new_df[['days_cube', 'days_sq', 'Days', 'Temp','Pop_dens_sq','Pop_dense', 'Population' ]]

# Get Final Response DataFrame
response = new_df[['Accumulated cases']]
```
