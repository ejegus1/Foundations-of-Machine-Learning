```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pandas as pd
%matplotlib inline
#%matplotlib notebook
```

```python
# data for a three-dimensional line

zline = np.linspace(0,15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)

# data for three-dimensional scatter plots
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
```

```python
from mpl_toolkits import mplot3d

fig = plt.figure(figsize=(15,15))


## first subplot 
ax1 = fig.add_subplot(1,2,1, projection='3d')
ax1.plot3D(xline, yline, zline, 'gray')
ax1.scatter3D(xdata, ydata, zdata, c=zdata, cmap ='Greens')

## second subplot
ax2 = fig.add_subplot(1,2,2, projection='3d')
ax2.plot3D(xline, yline, zline, 'gray')
ax2.scatter3D(xdata, ydata, zdata, c=zdata, cmap ='Greens')
```

change viewing angle with view_init method:

```python
fig = plt.figure(figsize=[15,15])

ax = fig.add_subplot(1,2,1, projection='3d')
ax.plot3D(xline, yline, zline, 'gray')
ax.view_init(25,35)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap ='Greens')
```
